```python
def naivebayies(df):
  temp={"sunnyp":0,"sunnyn":0,"overcastp":0,"overcastn":0,"rainp":0,"rainn":0,"hotp":0,"hotn":0,"mildp":0,"mildn":0,"coolp":0,"cooln":0,"highp":0,"highn":0,"normalp":0,"normaln":0,"Truep":0,"Truen":0,"Falsep":0,"Falsen":0}
  col=len(df.columns)
  pos=0
  neg=0
  for i in range(len(df.index)):
    for j in range(col-1):
      if(df.loc[i,"class"]=='p'):
         s=str(df.iloc[i,j])+'p'
         temp[s]+=1
         pos+=1
      else:
        s=str(df.iloc[i,j])+"n"
        temp[s]+=1
        neg+=1
  return temp,pos,neg;
lis,pos,neg=naivebayies(df)
print(lis)
print(pos/len(df.index))
```

```python
import pandas as pd
df=pd.read_csv("/content/lab4.csv")
df=pd.DataFrame(df)
df
```

```python
def vocab(df):
  l=[]
  for i in range(0,11):
    for j in (df.loc[i,"text"]).split():
      if j not in l:
        l.append(j)
  return l;
voc=vocab(df)
print(voc)
```

```python
from re import S
def count(temp,voc):
  l=[]
  s=0
  index=0
  for i in voc:
    c=0;
    if i in temp:
      for j in temp:
        if (j==i):
          c+=1
    s+=c
    l.append(c)
  return l,s;
def docs(df,voc):
  docpos={}
  docneg={}
  poscount=0
  negcount=0
  for i in range(0,11):
    temp=(df.loc[i,"text"]).split()
    lis,s=count(temp,voc)
    if df.loc[i,"class"]=="pos":
      docpos[i+1]=lis
      poscount+=s
    else:
      docneg[i+1]=lis
      negcount+=s
  return docpos,docneg,poscount,negcount;
docpos,docneg,poscount,negcount=docs(df,voc)
print(docpos,docneg)
print(poscount,negcount)
```

```python
def prior(voc,doc,count):
  dic={}
  for i in range(len(voc)):
    dic[voc[i]]=0
    for temp in doc.values():
      dic[voc[i]]+=temp[i]
    dic[voc[i]]=(dic[voc[i]]+1)/(count+len(voc))
  return dic
priprobpos=prior(voc,docpos,poscount)
priprobneg=prior(voc,docneg,negcount)
print(priprobpos)
print(priprobneg)
```

```python
def classcount(df):
  poscount=0
  negcount=0
  for i in range(0,11):
    if (df.loc[i,"class"])=="pos":
      poscount+=1
    else:
      negcount+=1
  return poscount,negcount;
def test(df,dicpos,dicneg,p,n):
  l=[]
  for i in range(10,len(df.index)):
    print(df.loc[i,"text"])
    pos=p
    neg=n
    for j in (df.loc[i,"text"]).split():
      if  j in dicpos.keys():
         pos*=dicpos[j]
      if  j in dicneg.keys():
         neg*=dicneg[j]
    if(pos>neg):
      l.append("pos")
    else:
      l.append("neg")
  return l;
p,n=classcount(df)
tes=test(df,priprobpos,priprobneg,p,n)
print(tes)
```

```python
def cal(df,l):
  c=0
  tp=0
  tn=0
  fp=0
  fn=0
  for i in range(10,len(df.index)):
    if(l[c]==df.loc[i,"class"]):
      if l[c]=="pos":
        tp+=1
      else:
        tn+=1
    else:
      if l[c]=="pos":
        fn+=1
      else:
        fp+=1
  return tp,tn,fp,fn;
tp,tn,fp,fn=cal(df,tes)
recall=tp/(tp+fn)
accuracy=(tp+tn)/(tp+tn+fp+fn)
precision=tp/(tp+fp)
print("recall=",recall," accuracy=",accuracy," precision=",precision)
```
