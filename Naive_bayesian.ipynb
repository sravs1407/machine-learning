```python
import pandas as pd
df=pd.read_csv("/content/p3.csv")
df=pd.DataFrame(df)
df
```

```python
def naivebayies(df):
  temp={"sunnyp":0,"sunnyn":0,"overcastp":0,"overcastn":0,"rainp":0,"rainn":0,"hotp":0,"hotn":0,"mildp":0,"mildn":0,"coolp":0,"cooln":0,"highp":0,"highn":0,"normalp":0,"normaln":0,"Truep":0,"Truen":0,"Falsep":0,"Falsen":0}
  col=len(df.columns)
  pos=0
  neg=0
  for i in range(len(df.index)):
    for j in range(col-1):
      if(df.loc[i,"class"]=='p'):
         s=str(df.iloc[i,j])+'p'
         temp[s]+=1
         pos+=1
      else:
        s=str(df.iloc[i,j])+"n"
        temp[s]+=1
        neg+=1
  return temp,pos,neg;
lis,pos,neg=naivebayies(df)
print(lis)
print(pos/len(df.index))
```

```python
testing=pd.read_csv("/content/testp3.csv")
testing
```

```python
def testingdata(test,pos,neg,temp,length):
  py=pos/length
  pn=neg/length
  print(py,pn)
  res=[]
  for i in range(len(test.index)):
    res1=py
    res2=pn
    for j in range(len(test.columns)-1):
      res1=res1*(temp[str(df.iloc[i,j])+'p']/pos)
      res2=res2*(temp[str(df.iloc[i,j])+'n']/neg)
    print(res1,res2)
    if(res1>res2):
      res.append('p')
    else:
      res.append('n')
  return res;
out=testingdata(testing,pos,neg,lis,len(df.index))
print(out)
```
